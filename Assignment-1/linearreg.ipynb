```python
def normal(X):    
    return (X-X.mean(axis=0))/X.std(axis=0)    
```

```python
df = pd.read_csv("H:\\BPHC\\FoDS\\Assignment-1\\3D_spatial_network.txt",sep=",",header=None)
t=df.iloc[:,3]
X=df.iloc[:,1:3]
t=np.c_[t]
X=normal(X)
t=normal(t)
X= np.c_[np.ones((len(X), 1)), X]
t_test=t[300000:]
X_test=X[300000:]
t_train=t[0:300000]
X_train=X[0:300000]
```

```python
class linearreg:
    def __init__(self, eta=0.01, n=1500,sr=1e-6):
        self.eta = eta
        self.n= n
        self.sr= sr
    def predict(self, x):   
        return np.dot(x, self.w_)
    def fit(self, x, y):
        self.cost_ = []
        self.rmse_ = []
        self.r2_ = []
        self.w_ = np.random.randn(x.shape[1], 1)
        m = x.shape[0]
        c1=c2=0
        for i in range(self.n):
            y_pred = np.dot(x, self.w_)
            residuals = y_pred - y
            gradient_vector = np.dot(x.T, residuals)
            k=(self.eta/m) * gradient_vector
            self.w_ -= (self.eta/m) * gradient_vector
            cost = np.sum((residuals ** 2)) / (2 * m)
            print(i, ' ',cost)
            c1=c2
            c2=cost
            self.rmse_.append((cost*2)**(0.5))
            ssr = np.sum((residuals)**2)
            sst = np.sum((y - np.mean(y))**2)
            self.r2_.append(1 - (ssr/sst))
            if abs(c2-c1)<self.sr:
                self.cost_.append(cost)
                break
            elif (c2-c1>=0)&(i>0):
                print('broken',c2)
                break    
            self.cost_.append(cost)
```

```python
A=linearreg() 
```
